```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
countries= gpd.read_file(r"data\countries.geojson")
clean_elections=pd.read_csv(r"data\clean_elections.csv")
unemployment=pd.read_csv(r"data\unemployment.csv")
gdpcapita=pd.read_csv(r"data\gdp_percapita.csv")
population=pd.read_csv(r"data\population.csv")
gini_index=pd.read_csv(r"data\gini_index.csv")
destination_ref=pd.read_csv(r"data\destination_refugees.csv")
origin_ref=pd.read_csv(r"data\origin_refugees.csv")
```

```python
year='2018'
temp=unemployment.loc[unemployment['Indicator Id']==3296, [year,'Country ISO3']]
temp.rename({year:'unemployment', 'Country ISO3':'ISO_A3'}, axis=1, inplace=True)
result=pd.merge(countries,temp,how='left', on='ISO_A3')
temp=clean_elections.loc[clean_elections['Indicator Id']==41828, ['Country ISO3',year]]
temp.rename({year:'clean_elections', 'Country ISO3':'ISO_A3'}, axis=1, inplace=True)
result=pd.merge(result,temp,how='left', on='ISO_A3')
temp=gdpcapita.loc[:,['Country Code',year]]
temp.rename({year:'gdp_capita', 'Country Code':'ISO_A3'}, axis=1, inplace=True)
result=pd.merge(result,temp,how='left', on='ISO_A3')
temp=population.loc[:,['Country Code',year]]
temp.rename({year:'population', 'Country Code':'ISO_A3'}, axis=1, inplace=True)
result=pd.merge(result,temp,how='left', on='ISO_A3')
temp=gini_index.loc[:,['Country Code',year]]
temp.rename({year:'gini_index', 'Country Code':'ISO_A3'}, axis=1, inplace=True)
result=pd.merge(result,temp,how='left', on='ISO_A3')
temp=destination_ref.loc[:,pd.concat([pd.Series(['Country Code']), pd.Series(range(int(year)-4,int(year)+2)).astype(str)])]
temp['avg_inc_ref_5y'] = temp[pd.Series(range(int(year)-4,int(year)+1)).astype(str)].sum(axis=1)/5
temp['ratio_inc_ref_5y'] = temp[str(int(year))]/temp[str(int(year)-4)]
temp.rename({year:'incomming_ref', 'Country Code':'ISO_A3', str(int(year)+1):'incomming_ref_target'}, axis=1, inplace=True)
result=pd.merge(result,temp.loc[:,['ISO_A3', 'incomming_ref', 'avg_inc_ref_5y', 'ratio_inc_ref_5y', 'incomming_ref_target']],how='left', on='ISO_A3')
result
```

```python
countries_geom = result.geometry.to_crs(4326)#"ESRI:54009"
fig, ax = plt.subplots(ncols=1, nrows=1)
countries_geom.plot(ax=ax, color='green', zorder=1)
```

```python
countries_geom = result.geometry.to_crs("ESRI:54009")#
fig, ax = plt.subplots(ncols=1, nrows=1)
countries_geom.plot(ax=ax, color='green', zorder=1)
```

```python
#countries_buffer = countries_geom.buffer(200000)
```

```python
# add NEIGHBORS column
result["neighbors"] = None  

for index, country in result.iterrows():   

    # get 'not disjoint' countries
    neighbors = result[~result.geometry.disjoint(country.geometry)].ISO_A3.tolist()

    # remove own name of the country from the list
    neighbors = [ name for name in neighbors if country.ISO_A3 != name ]

    # add names of neighbors as NEIGHBORS value
    result.at[index, "neighbors"] = ", ".join(neighbors)
```

```python
sum_out_ref_neighbors = []
temp = origin_ref.loc[:,['Country Code', str(int(year))]]
for country in result.ISO_A3:
    neigh_ref = temp[temp["Country Code"].isin([x.strip(' ') for x in result[result.ISO_A3 == country].neighbors.str.split(',').reset_index(drop=True)[0]])]
    sum_out_ref_neighbors.append(neigh_ref[str(int(year))].sum())
```

```python
result['outgoing_ref_neighbors'] = sum_out_ref_neighbors
```

```python
final2 = gpd.GeoDataFrame(result, geometry=countries.geometry)
```

```python
final2.crs.axis_info[0].unit_name
```

```python
prova = final2.to_crs(epsg=4087)
```

```python
fig, ax = plt.subplots(ncols=1, nrows=1)
prova.plot(ax=ax, color='green', zorder=1)
```

```python
import geopy.distance
final2['Centroid']=prova.geometry.centroid.to_crs(4326)
coord_list = [(x,y) for x,y in zip(final2['Centroid'].x , final2['Centroid'].y)]
#coord_list = [(y,x) for y,x in zip(final2['Centroid'].y , final2['Centroid'].x)]
dist=[]
for cent in coord_list:
    ind=[]
    for cent2 in coord_list:
        ind.append(geopy.distance.distance(cent, cent2).km)
    dist.append(ind)
sdist=pd.Series(dist)
final2['Distance']=sdist
#final2['Centroid']=final2.Centroid.to_crs(4326)
```

```python
### Weight matrix
import numpy as np
# inverse distance weight matrix
Wdb=np.zeros([255,255])
# filling matrix
for i in range(254):
    for j in range(254):
        if i!=j:
            Wdb[i][j]=1/(final2.Distance[i][j])
        else: continue
```

```python
temp = pd.merge(final2,
               origin_ref.loc[:,['Country Code', str(int(year))]],
               how='left', left_on='ISO_A3', right_on='Country Code')
```

```python
final2['outgoing_ref_other_weighted'] = pd.Series(Wdb.dot(np.where(np.isnan(temp[str(int(year))]), 0, temp[str(int(year))])))
```

```python
final2.to_csv(r"data\final_dataset.csv",index=False)
```
